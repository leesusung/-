```python
stock = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/SK.csv') 
stock.head()
```

```python
stock = stock[['Date','Close','Volume']]
```

```python
#  기준일포함 과거3일 평균
stock['Close_MA_3'] = stock['Close'].rolling(3).mean()
```

```python
stock['Close_ma_3_lag1'] =stock['Close_MA_3'].shift(1)
```

```python
stock
```

```python
stock['close_ma_3(2)'] = stock['Close'].rolling(3,min_periods=0).mean()
```

```python
stock
```

```python
stock = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/sknetworks.csv', sep=',', skipinitialspace=True) 
stock = stock[['Date', 'Close', 'Volume']]
stock.head()
```

```python
stock['1dayago'] =stock['Close'].shift()
stock['2dayago'] =stock['Close'].shift(2)
```

```python
stock
```

```python
stock['diff']= stock['1dayago']-stock['2dayago'] 
```

```python
stock.loc[stock['diff'] > 0 ,'diff2' ] =1
stock.loc[stock['diff'] < 0 ,'diff2' ] =0
```

```python
stock
```

```python
#  q4
stock['max_3day']= stock['Close'].rolling(3).max()
```

```python
stock
```

```python
# q5
stock['max_3day_lag'] = stock['max_3day'].shift()
```

```python
# q6
stock['volume_7day_lag'] = stock['Volume'].rolling(7).mean().shift()
```

```python
stock.head(8)
```

```python
# q7
stock['vol_7day_ver2'] = stock['Volume'].rolling(7, min_periods=1).mean().shift()
```

```python
stock['vol_7day_ver2'] = stock['Volume'].rolling(7, min_periods=1).mean()
```
